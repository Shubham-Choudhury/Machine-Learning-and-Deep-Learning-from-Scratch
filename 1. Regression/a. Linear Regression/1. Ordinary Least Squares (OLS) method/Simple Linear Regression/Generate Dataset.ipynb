```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Set the random seed for reproducibility

```python
np.random.seed(42)
```

### Parameters

```python
n = 1000
size_min, size_max = 500, 10000  # Size in square feet
price_per_sqft = 156            # Price per square foot in dollars
noise_std_dev = 99999         # Standard deviation of noise
```

### Generate house sizes

```python
sizes = np.random.uniform(size_min, size_max, n)
```

### Generate prices with some noise

```python
prices = sizes * price_per_sqft + np.random.normal(0, noise_std_dev, n)
```

### Round sizes and prices to 3 decimal points

```python
sizes = np.round(sizes, 3)
prices = np.round(prices, 3)
```

### Create a DataFrame

```python
data = pd.DataFrame({'Size (sqft)': sizes, 'Price': prices})
```

### Display first few rows of the dataset

```python
print(data.head())
```

### Save to a CSV file

```python
data.to_csv('house_price_prediction_dataset.csv', index=False)
```

### Plot the dataset

```python
plt.scatter(data['Size (sqft)'], data['Price'], alpha=0.5)
plt.title('House Price Prediction Dataset')
plt.xlabel('Size (sqft)')
plt.ylabel('Price ($)')
plt.show()
```
